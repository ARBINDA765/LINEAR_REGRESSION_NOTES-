```python
#Importing the packages
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
#Reading the dataset 
df_adv=pd.read_csv("Advertising.csv",index_col=0)
df_adv.head()
```

```python
#Independent Features
X =df_adv[["TV", "Radio","Newspaper"]]
#Dependent Features
y = df_adv['Sales']
```

```python
#Adding the constrant to the X 
X = sm.add_constant(X)
```

```python
X
```

```python
## fit a OLS model with intercept on TV and Radio
model=sm.OLS(y,X).fit()
```

```python
#Getting the summary
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```
